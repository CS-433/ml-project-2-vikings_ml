This notebook contains the functions need for running our traditional UNet.

```python
from Unet import UNet
from tensorflow import keras
from data_handling_unet import create_data_model
import os
import matplotlib.image as mpimg
import random
import numpy as np
import math
from tensorflow import keras
```

```python
# Load Unet NN architecture
model = UNet().get_model(image_shape=(256,256,3))
model.summary()
```

```python
# Creating data model. 
train_gen, val_gen = create_data_model()

# Training and saving the best model
epochs = 20
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])
callbacks = [keras.callbacks.ModelCheckpoint("Unet_roadsegment4.h5", save_best_only=True)]
model.fit(train_gen, epochs=epochs, validation_data=val_gen, callbacks=callbacks)
```
